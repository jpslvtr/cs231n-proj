```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Load and display some example frames from scene1
scene_path = "data/processed/scene1"
frame_files = [f for f in os.listdir(scene_path) if f.endswith(".png")]

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for i, frame_file in enumerate(frame_files[:5]):
    frame_path = os.path.join(scene_path, frame_file)
    image = Image.open(frame_path)
    axes[i].imshow(image)
    axes[i].set_title(f"Frame {i+1}")
    axes[i].axis('off')
plt.show()
```
